```
m1<-lm(math4~lexppp+lenroll+lunch+l2)
summary(m1)
m1
```

$R^{2}$= 0.379, $n$ = 1823

$lunch^{2}$ is not statistically significant at the 5% significance level. Every other variable is statistically significant at the 0.001 level

```
install.packages("quantreg")
library(quantreg)
```

```
fit1= summary(rq(math4~lexppp+lenroll+lunch+l2, tau= c(.1,.2,.3,.4,.5,.6,.7,.8,.9)))
summary(fit1)
fit1
plot(fit1, mfrow= c(1,2))
#plot(summary(fit1))
```

$lenroll$ is not statistically significant at 5% level at tau 0.1

$lexppp$ is not statistically significant at 5% level at tau 0.9

$lunch^{2}$ is not statistically  significant level across all quantiles

From the figures plotted, the plot on luch shows the comparison between OLS and Quantile regression. The coefficient of $lunch$ in OLS is $-0.4414$, which denotes that on average one unit decrease in $lunch$ will lead to 0.4414 decrease in math4 with all the other varibles held constant.

In quantile regression, the coefficient of $lunch$ varies across the quantiles and the absolute value of the coefficient is higher especially in the lower quantile.

$lunch$ is significant at any quantile, but $lunchsq$ is not significant across any quantiles. But on observation, the coefficient of $lunchsq$ shares the same negative sign with $lunch$ in quantiles 0.2 to 0.7.

We could tell that the quantile regression will provide a more complete picture between $lunch$ and $math4$ because it gives us more detailed information.
